```python
import numpy as np
import matplotlib.pyplot as plt
ages = [36, 47, 29, 82, 94, 54, 23, 98, 10, 50, 34, 76, 53, 47, 81, 73]
range = (0, 100)
bins = 10
plt.hist(ages, bins, range, color='purple', histtype="bar", rwidth=0.5)
plt.xlabel("Ages")
plt.ylabel("Bins")
plt.title("Histogram Plot")
plt.show()
```
