```python
import numpy as np
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [52, 67, 93, 47, 56]
tick_labels = ['one', 'two', 'three', 'four', 'five']
plt.bar(x, y, tick_label=tick_labels, width=0.5, color=['purple'])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Bar Graph")
plt.show()
```
